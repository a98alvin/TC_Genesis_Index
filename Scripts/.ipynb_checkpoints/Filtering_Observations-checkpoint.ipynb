```python
train_labels_predropped_read = xr.open_dataset("/glade/work/acheung/Initial_RF_Datasets/train_labels_predropped.nc")
train_labels_predropped = train_labels_predropped_read.stack(z=("Month","Latitude","Longitude","Year"))

test_labels_predropped_read = xr.open_dataset("/glade/work/acheung/Initial_RF_Datasets/test_labels_predropped.nc")
test_labels_predropped = test_labels_predropped_read.stack(z=("Month","Latitude","Longitude","Year"))
```

```python
month_label = train_labels_predropped_read.sel(Month=6).to_array()[0]
mean_month_label = np.mean(month_label,axis=2)
filtered_month = gaussian_filter(mean_month_label, sigma=1.2)
filtered_month_with_coords = xr.DataArray(filtered_month, coords=[mean_month_label.coords['Latitude'],mean_month_label.coords['Longitude']])

plt.figure()
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_global()
ax.coastlines()
ax.gridlines(draw_labels=True)
filtered_month_plot = ax.pcolormesh(filtered_month_with_coords.coords['Longitude'],
                                    filtered_month_with_coords.coords['Latitude'],
                                    filtered_month_with_coords,transform=ccrs.PlateCarree(),vmin=0)
ax.set_extent([-100, 0, 0, 80])
cbar = plt.colorbar(filtered_month_plot)
```
