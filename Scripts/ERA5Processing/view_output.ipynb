```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
vort = xr.open_dataset('/glade/scratch/acheung/abs_vort/e5.oper.an.pl.128_138_vo.ll025sc.7d_mean.1940093000_1940100623.nc')
```

```python
plt.figure()
ax1 = plt.axes(projection=ccrs.PlateCarree())
ax1.set_global()
ax1.coastlines()
ax1.gridlines(draw_labels=True)
ax1.set_extent([-100, 0, 0, 80])
vort['Absolute Vorticity'].plot(vmin=0,vmax=0.00015)
```
